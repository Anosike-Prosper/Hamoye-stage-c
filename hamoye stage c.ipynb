```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix
```

```python
data= pd.read_csv('Data_for_UCI_named (1).csv')

data
```

```python
data.isnull().sum()
```

```python
data.drop('stab', axis=1, inplace=True)

data
```

```python
data.stabf.value_counts()
```

```python
x = data.drop('stabf', axis= 1)

y= data['stabf']
```

```python

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size= 0.2 , random_state= 1 ) # splitting the data
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
scaler = StandardScaler() # using the standard scaler

```

```python
normalised_train_df = scaler.fit_transform(X_train) 
normalised_train_df = pd.DataFrame(normalised_train_df, columns=X_train.columns )

normalised_train_df.head()
```

```python
X_test= X_test.reset_index(drop= True)

normalised_test_df = scaler.transform(X_test) 
normalised_test_df = pd.DataFrame(normalised_test_df, columns=X_train.columns )

normalised_test_df.head()
```

```python
# random forest classifier

from sklearn.ensemble import RandomForestClassifier


Random_forest = RandomForestClassifier(random_state=1)
Random_forest.fit(normalised_train_df, y_train)

prediction = Random_forest.predict(normalised_test_df)

```

```python
print('Accuracy score is', accuracy_score(y_test, prediction)) 
print('Precision score for label stable is' ,  precision_score(y_test, prediction, pos_label='stable'))
print('Recall score for label stable is', recall_score(y_test, prediction, pos_label='stable'))
print('F1 score is',  f1_score(y_test, prediction, pos_label='stable'))
```

```python
from xgboost import XGBClassifier

xgb_class = XGBClassifier(random_state =1, max_depth=3, learning_rate= 0.1)
xgb_class.fit(normalised_train_df, y_train)


xgb_pred = xgb_class.predict(normalised_test_df)


```

```python
round(accuracy_score(y_true=y_test, y_pred=xgb_pred), 4)
```

```python
# for this, my answer was not in the option, so i chose the closest answer

from lightgbm import LGBMClassifier


lgbm_class = LGBMClassifier(random_state=1)
lgbm_class.fit(normalised_train_df, y_train)
lgbm_predict  = lgbm_class.predict(normalised_test_df)

round(accuracy_score(y_test, lgbm_predict) , 4)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import RandomizedSearchCV

Ext_tree = ExtraTreesClassifier (random_state = 1)  
```

```python
n_estimators = [50, 100, 300, 500, 1000]    # Given we are to use this parameters to answer the question

min_samples_split = [2, 3, 5, 7, 9]

min_samples_leaf = [1, 2, 4, 6, 8]

max_features = ['auto', 'sqrt', 'log2', None] 

hyperparameter_grid = {'n_estimators': n_estimators,

                       'min_samples_leaf': min_samples_leaf,

                       'min_samples_split': min_samples_split,

                       'max_features': max_features}
```

```python
Random_search = RandomizedSearchCV(estimator = Ext_tree, param_distributions= hyperparameter_grid, random_state=1,cv = 5, n_iter=10,scoring='accuracy',n_jobs=1, verbose=1)
```

```python
hyp_params = Random_search.fit(normalised_train_df, y_train)
```

```python
hyp_params.best_params_ # these are the best hyperparameters for the model
```

```python
# new model using hyper parameters above

New_ext_tree = ExtraTreesClassifier(n_estimators=1000, min_samples_split=2, 
                                 min_samples_leaf=8, max_features=None)
New_ext_tree.fit(normalised_train_df, y_train)
Ext_pred = New_ext_tree.predict(normalised_test_df)
```

```python
accuracy_score(y_test, Ext_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, Ext_pred))
```
